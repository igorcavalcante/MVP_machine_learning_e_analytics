```python
from astropy.time import Time

import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler

import tensorflow as tf

from keras.optimizers import Adam
from keras.optimizers.schedules import ExponentialDecay
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense, Activation, Flatten, Dropout

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('alpha_ori_2000_2025.csv')

# Os dados completos vindos da AAVSO
data

# Analisando a incerteza
# data.sort_values(by="Uncertainty", ascending=False)
```

```python
# Só precisarei da data, magnitude e incerteza
data = data.iloc[:, [0,1,2]]
data.head()
```

```python
# mudando a data para o calendário gregoriano
data.loc[:, "DataHora"] = data["JD"].apply(lambda d: Time(d, format='jd').iso)

# transformando os valores que são NaN na incerteza em 0.0, pois são os menores valores possíveis
data["Uncertainty"] = data["Uncertainty"].fillna(0.0)

data.head()
```

```python
# mudando a magnitude pra float
data["Magnitude"] = data["Magnitude"].str.replace("<", "", regex=False).astype(float)
```

```python
# analisando a incerteza para ver como posso descartar as entradas duplicadas  no mesmo dia
# Percebe-se que há dados  com até 60% de incerteza
print(data["Uncertainty"].describe())
```

```python
# Quero somente uma medição por dia
# Utilizarei sempre a medição com menor incerteza daquele dia
# data.loc[data.groupby[data["data"].dt.date["Uncertainty"].idxmin()]
# data.groupby["Data"]


data["Dia"] = pd.to_datetime(data["DataHora"]).dt.date
grouped = data.groupby("Dia")
grouped
idx = grouped["Uncertainty"].idxmin()

byDay = data.loc[idx].sort_values("Dia").reset_index(drop=True)

# O dado com maior incerteza agora é de 8% e temos somente uma entrada por dia
# A média e desvio padrão também caíram muito
print(byDay["Uncertainty"].describe())
print(f"Total de dias presentes: {len(byDay)}")
```

```python
# Verificando quantos dias não tiveram medições
allDays = pd.date_range(
    start=byDay["Dia"].min(),
    end=byDay["Dia"].max(),
    freq="D"
)

faltantes = allDays.difference(byDay["Dia"].unique())

print(f"Total de dias faltantes: {len(faltantes)}")
print(faltantes[:20])  # mostrar só os primeiros 20
```

```python
# Completando com o valor da medição seguinte para os dias que não tiveram medições
byDayComplete = byDay.set_index("Dia").reindex(allDays).rename_axis("Dia").reset_index()
byDayComplete = byDayComplete.ffill()

byDayComplete
```

```python
# Agora  com a incerteza mais baixa e apenas uma medição por dia eu posso gerar as janelas temporais
mmscaler = MinMaxScaler((-1.0, 1.0))
Y = mmscaler.fit_transform(byDayComplete[["Magnitude"]])

# colocando um número divisível por 30 pra simplificar
Y = Y[:8760]
```

```python
nb_samples= 8760
nb_test_samples=780
sequence_length = 10

X_ts = np.zeros(shape=(nb_samples - sequence_length, sequence_length, 1), dtype=np.float32)
Y_ts = np.zeros(shape=(nb_samples - sequence_length, 1), dtype=np.float32)

for i in range(0, nb_samples - sequence_length):
    X_ts[i] = Y[i:i+sequence_length]
    Y_ts[i] = Y[i + sequence_length]

X_ts_train = X_ts[0:nb_samples - nb_test_samples, :]
Y_ts_train = Y_ts[0:nb_samples - nb_test_samples]

X_ts_test = X_ts[nb_samples - nb_test_samples:, :]
Y_ts_test = Y_ts[nb_samples - nb_test_samples:]
```

```python
model = Sequential()

model.add(Input(batch_size=sequence_length, shape=(sequence_length, 1)))
model.add(LSTM(8, stateful=True))
model.add(Dropout(0.2))
model.add(LSTM(8, stateful=True))  # última sem return_sequences
model.add(Dropout(0.2))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="linear"))
```

```python
#optimizer = Adam(learning_rate=lr_schedule)

#model.compile(
#    loss="mse",
#    optimizer=optimizer,
#    metrics=[tf.keras.metrics.MeanAbsoluteError(),
#             tf.keras.metrics.RootMeanSquaredError(),
#             tf.keras.metrics.MeanAbsolutePercentageError()]
#)


lr_schedule = ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=10000,
    decay_rate=0.0001
)

model.compile(optimizer=Adam(learning_rate=lr_schedule), #treinada com taxa de aprendizado 0.001
              loss='mse', # e função de perda/métrica MSE
              metrics=[
                  tf.keras.metrics.MeanAbsoluteError(),
                  tf.keras.metrics.MeanAbsolutePercentageError()
              ])
```

```python
history = model.fit(X_ts_train, Y_ts_train,
          batch_size=30, #para lotes de 20 em 20
          epochs=50, # por 50 épocas
          shuffle=False, #sem embaralhar os dados de treinamento a cada época
          validation_data=(X_ts_test, Y_ts_test)) #fazendo a validação com o conjunto separado para testes
```

```python
loss, mae, rmse, mape = model.evaluate(X_ts_test, Y_ts_test, verbose=0, batch_size=30)
print(f"MSE: {loss:.4f} | MAE: {mae:.4f} | RMSE: {rmse:.4f} | MAPE: {mape:.2f}%")
```

```python
# lista todos os dados contidos no histórico de treinamento
#print(history.history.keys())

# plotamos o erro calculado a partir do histórico
plt.plot(history.history['mse'])  # erro médio quadrático (treinamento)
plt.plot(history.history['val_mse']) # erro médio quadrático (teste)
plt.title('MSE do Modelo')
plt.ylabel('mse')
plt.xlabel('época')
plt.legend(['treino', 'teste'], loc='upper left')
plt.show()

# plotamos o erro/perda a partir do histórico
plt.plot(history.history['loss']) # perda/erro durante o treinamento
plt.plot(history.history['val_loss']) #perda/erro durante a validação (teste)
plt.title('Perda do modelo')
plt.ylabel('perda')
plt.xlabel('época')
plt.legend(['treino', 'teste'], loc='upper left')
plt.show()
```

```python
# Agora, vamos visualizar os dados reais vs os dados calculados na predição (usando os dados de treinamento)
plt.plot(Y_ts_train) #plota dados originais
plt.plot(model.predict(X_ts_train, batch_size=30)) #plota dados resultantes
plt.legend(['Dado Real', 'Predição'], loc='upper left')
plt.show()
```

```python
# e vamos visualizar os dados reais vs os dados calculados na predição (usando os dados de validação)
plt.plot(Y_ts_test)
plt.plot(model.predict(X_ts_test, batch_size=30))
plt.legend(['Dado Real', 'Predição'], loc='upper left')
plt.show()
```
